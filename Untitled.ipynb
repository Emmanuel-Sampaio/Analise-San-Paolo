```python


# Seu df já filtrado e com pedidos válidos
df_validos = df[df['Total do Pedido'] > 0].copy()

# Função para extrair a cidade da string da loja
def extrair_cidade(loja):
    prefixo = 'SAN PAOLO GELATO & CAFÉ '
    if loja.startswith(prefixo):
        return loja[len(prefixo):].split('|')[0].strip()
    else:
        return None

df_validos['Cidade'] = df_validos['Loja'].apply(extrair_cidade)

# Dicionário de cidades para estados (com base no que você passou)
mapa_cidade_estado = {
    'FORTALEZA': 'CE',
    'RECIFE': 'PE',
    'SALVADOR': 'BA',
    'TERESINA': 'PI',
    'CAMPINA GRANDE': 'PB',
    'JOÃO PESSOA': 'PB',
    'NATAL': 'RN',
    'BELÉM': 'PA',
    'JUAZEIRO DO NORTE': 'CE',
    'MOSSORÓ': 'RN',
    'ARACAJU': 'SE',
    'SÃO LUÍS': 'MA',
    'PERNAMBUCO': 'PE',
    'SÃO PAULO': 'SP',
    'MANAUS': 'AM',
    # Adicione mais se precisar
}

# Mapear cidade para estado
df_validos['Estado'] = df_validos['Cidade'].str.upper().map(mapa_cidade_estado)



# Agrupar lucro por estado
lucro_por_estado = df_validos.groupby('Estado')['Total do Pedido'].sum().sort_values(ascending=False)

print(lucro_por_estado)

# Opcional: plotar o lucro por estado
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
lucro_por_estado.plot(kind='bar', color='green')
plt.title('Lucro por Estado')
plt.ylabel('Lucro Total (R$)')
plt.xlabel('Estado')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
lojas = df_validos.groupby('Loja').agg({
    'Total do Pedido': ['sum', 'count', 'mean']
}).reset_index()

lojas.columns = ['Loja', 'Total de Vendas', 'Número de Pedidos', 'Ticket Médio']

# Ordena pelo Ticket Médio
lojas = lojas.sort_values(by='Ticket Médio', ascending=False)

# Normaliza o nome da loja, removendo o prefixo e espaços extras
lojas['Loja_Normalizada'] = lojas['Loja'].str.replace('SAN PAOLO GELATO & CAFÉ', '', regex=False).str.strip()

# Plotando o gráfico usando a coluna normalizada
plt.figure(figsize=(12, 8))
plt.barh(lojas['Loja_Normalizada'], lojas['Ticket Médio'], color='skyblue')
plt.xlabel('Ticket Médio (R$)')
plt.title('Ticket Médio por Loja')
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```

Podemos ver aqui que o ticket médio por loja é bem parecido, sem alguma loja que destoe muito, o que nos leva a crer que o que muda mesmo é a frequência de vendas e não o quanto os clientes gastam na loja

```python
# faturamento total por estado
faturamento_total_por_estado = df_validos.groupby('Estado')['Total do Pedido'].sum()

# número de lojas únicas por estado
num_lojas_por_estado = df_validos.groupby('Estado')['Loja'].nunique()

# faturamento médio por loja em cada estado
faturamento_medio_por_estado = faturamento_total_por_estado / num_lojas_por_estado
print(faturamento_medio_por_estado)

plt.figure(figsize=(10, 6))
faturamento_medio_por_estado.sort_values(ascending=False).plot(kind='bar', color='coral')
plt.title('Faturamento Médio por Loja por Estado')
plt.xlabel('Estado')
plt.ylabel('Faturamento Médio (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
df['AnoMes'] = df['Data do Pedido'].dt.to_period('M')
lucro_por_mes = df.groupby('AnoMes')['Total do Pedido'].sum()
print(lucro_por_mes)



plt.figure(figsize=(10, 6))
lucro_por_mes.sort_values(ascending=False).plot(kind='bar', color='coral')
plt.title('Faturamento por Mês')
plt.xlabel('Estado')
plt.ylabel('Faturamento Médio (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Cria uma lista com todos os meses existentes no dataframe (únicos)
meses = sorted(df['Data do Pedido'].dt.month.unique())

# Configura a figura com subplots (por exemplo, 2 linhas x 3 colunas para 6 meses)
fig, axs = plt.subplots(2, 3, figsize=(15, 8), constrained_layout=True)

# Achata a matriz de axs para iterar facilmente
axs = axs.flatten()

for i, mes in enumerate(meses):
    # Filtra só o mês atual
    df_mes = df[df['Data do Pedido'].dt.month == mes]
    
    # Agrupa por dia no mês
    lucro_por_dia = df_mes.groupby(df_mes['Data do Pedido'].dt.to_period('D'))['Total do Pedido'].sum()
    
    # Converte o índice para string para melhorar o plot
    dias = lucro_por_dia.index.astype(str)
    valores = lucro_por_dia.values
    
    # Plota no subplot correspondente
    axs[i].plot(dias, valores, marker='o', linestyle='-')
    axs[i].set_title(f'Mês {mes}')
    axs[i].tick_params(axis='x', rotation=45)
    axs[i].set_xlabel('Dia')
    axs[i].set_ylabel('Lucro')
    
    # Ajusta para mostrar menos ticks no eixo x (opcional)
    if len(dias) > 10:
        axs[i].set_xticks(dias[::max(1, len(dias)//10)])

# Remove subplots vazios caso tenha menos de 6 meses
for j in range(len(meses), len(axs)):
    fig.delaxes(axs[j])

plt.suptitle('Lucro Diário por Mês')
plt.show()
```

```python
# Agrupar por dia (usando a coluna de data convertida para datetime)
lucro_por_dia = df.groupby(df['Data do Pedido'].dt.date)['Total do Pedido'].sum()

# Dia que mais lucrou
dia_mais_lucro = lucro_por_dia.idxmax()
valor_mais_lucro = lucro_por_dia.max()

# Dia que menos lucrou
dia_menos_lucro = lucro_por_dia.idxmin()
valor_menos_lucro = lucro_por_dia.min()

# Lucro médio diário
lucro_medio_diario = lucro_por_dia.mean()
mediana_lucro_diario = lucro_por_dia.median()


print(f"Mediana dos lucros {mediana_lucro_diario:.2f}")
print(f'Dia que mais lucrou: {dia_mais_lucro} com lucro de {valor_mais_lucro:.2f}')
print(f'Dia que menos lucrou: {dia_menos_lucro} com lucro de {valor_menos_lucro:.2f}')
print(f'Lucro médio diário: {lucro_medio_diario:.2f}')
print(f"{lucro_por_dia.std()}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt


contagem_pagamentos = df['Forma de pagamento'].value_counts(normalize=True)

limite = 0.02  # 2%

principais = contagem_pagamentos[contagem_pagamentos >= limite]
outros = contagem_pagamentos[contagem_pagamentos < limite].sum()

# Juntar principais + "Outros" usando concat
contagem_agrupada = pd.concat([principais, pd.Series({'Outros': outros})])

# Plotar gráfico de pizza
plt.figure(figsize=(8,8))
contagem_agrupada.plot.pie(autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Porcentagem das Formas de Pagamento (Agrupado)')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Converter a coluna Data Pedido para datetime, dia primeiro
df['Data Pedido'] = pd.to_datetime(df['Data Pedido'], dayfirst=True)

# Criar coluna com o nome do dia da semana em inglês
df['DiaSemana'] = df['Data Pedido'].dt.day_name()

# Mapear dias para português
mapa_dias = {
    'Monday': 'Segunda-feira',
    'Tuesday': 'Terça-feira',
    'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira',
    'Friday': 'Sexta-feira',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo'
}
df['DiaSemana'] = df['DiaSemana'].map(mapa_dias)

# Agrupar o total do pedido por dia da semana
lucro_por_dia = df.groupby('DiaSemana')['Total do Pedido'].sum()

# Para manter a ordem correta dos dias da semana, vamos ordenar explicitamente:
ordem_dias = ['Segunda-feira', 'Terça-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sábado', 'Domingo']
lucro_por_dia = lucro_por_dia.reindex(ordem_dias)

# Plotar o gráfico
plt.figure(figsize=(10,6))
bars = plt.bar(lucro_por_dia.index, lucro_por_dia.values, color='teal')
plt.title('Lucro Total por Dia da Semana', fontsize=16)
plt.xlabel('Dia da Semana', fontsize=14)
plt.ylabel('Lucro Total (R$)', fontsize=14)
plt.xticks(rotation=45)

# Adicionar valores em cima das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + max(lucro_por_dia.values)*0.01, f'{yval:,.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Garantir que 'Data Pedido' esteja em datetime
df['Data Pedido'] = pd.to_datetime(df['Data Pedido'], errors='coerce')

# Extrair quantidade e nome do produto
df[['Quantidade', 'Produto']] = df['Detalhes'].str.extract(r'^(\d+)\s+(.*)')
df['Quantidade'] = pd.to_numeric(df['Quantidade'], errors='coerce')

# Criar coluna com o dia da semana em português (ex: segunda-feira)
df['Dia da Semana'] = df['Data Pedido'].dt.day_name(locale='pt_BR').str.lower()

# Agrupar por dia da semana e produto
agrupado = df.groupby(['Dia da Semana', 'Produto'])['Quantidade'].sum().reset_index()

# Pegar o produto mais vendido por dia
mais_vendidos = (
    agrupado.sort_values(['Dia da Semana', 'Quantidade'], ascending=[True, False])
    .groupby('Dia da Semana')
    .first()
    .reset_index()
)

# Ordenar dias da semana
ordem_dias = ['segunda-feira', 'terça-feira', 'quarta-feira', 'quinta-feira', 'sexta-feira', 'sábado', 'domingo']
mais_vendidos['Dia da Semana'] = pd.Categorical(mais_vendidos['Dia da Semana'], categories=ordem_dias, ordered=True)
mais_vendidos = mais_vendidos.sort_values('Dia da Semana')

# Exibir no formato desejado
for _, row in mais_vendidos.iterrows():
    print(f"{row['Dia da Semana']} - {row['Produto']}")
```

```python
import pandas as pd



# Criar coluna com o dia da semana
df["dia_semana"] = df["Data Pedido"].dt.day_name()

# Criar coluna indicando se é fim de semana
df["fim_de_semana"] = df["dia_semana"].isin(["Saturday", "Sunday"])

# Criar coluna binária: usou voucher?
df["usou_voucher"] = df["Bonificação"].notnull() & (df["Bonificação"] != 0)

# Agrupar por dia da semana e uso de voucher
resumo = df.groupby(["fim_de_semana", "usou_voucher"]).agg({
    "N°/pedido": "count"
}).reset_index().rename(columns={"N°/pedido": "quantidade_pedidos"})

print(resumo)


# Filtros
fds_voucher = resumo[(resumo["fim_de_semana"] == True) & (resumo["usou_voucher"] == True)]["quantidade_pedidos"].values[0]
fds_sem_voucher = resumo[(resumo["fim_de_semana"] == True) & (resumo["usou_voucher"] == False)]["quantidade_pedidos"].values[0]

if fds_voucher > fds_sem_voucher:
    print("🔎 Maior volume de vendas nos fins de semana com uso de vouchers.")
else:
    print("📉 Vouchers não parecem aumentar a quantidade de vendas nos fins de semana.")
```
